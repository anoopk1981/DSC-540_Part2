#### a) Generate 50 data points from this function in the range [â€“ 3, 3]

```python
import numpy as np
import matplotlib.pyplot as plt

data=np.linspace(-3, 3, 50)
sig_data=np.sinc(data)

plt.plot(data, sig_data, marker = "x")
plt.title("sinc()")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#### b) Add Gaussian noise to the data.

```python
noise_data = np.random.normal(0,0.5,50)
with_noise_data=sig_data+noise_data

plt.plot(data,with_noise_data )
plt.title("sinc() with noise")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#### c) Train an SVM regressor with the data generated in (a). Define (and explain) suitable parameters required for training the regressor.

###### Support Vector Regression (SVR) uses the same principle as SVM, but for regression problems.
###### The problem of regression is to find a function that approximates mapping from an input domain to real numbers on the basis of a training sample.
###### We can think of Support Vector Regression as the counterpart of SVM for regression problems. SVR acknowledges the presence of non-linearity in the data and provides a proficient prediction model.
###### Consider these two red lines as the decision boundary and the green line as the hyperplane. Our objective, when we are moving on with SVR, is to basically consider the points that are within the decision boundary line.
